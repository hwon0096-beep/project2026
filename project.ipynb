```python
orders["delivery_days"] = (
    orders["order_delivered_customer_date"]
    - orders["order_purchase_timestamp"]
).dt.days
```

```python
orders["approval_hours"] = (
    orders["order_approved_at"]
    - orders["order_purchase_timestamp"]
).dt.total_seconds() / 3600
```

```python
orders["delivery_delay_days"] = (
    orders["order_delivered_customer_date"]
    - orders["order_estimated_delivery_date"]
).dt.days
```

## Items

The order_items table serves as the central transaction fact table. Each record represents an individual product purchased within an order and contains product, seller, price, and freight information. The table enables revenue analysis, seller performance evaluation, product category analysis, and customer spending calculations.

```python
items["shipping_limit_date"] = pd.to_datetime(
    items["shipping_limit_date"]
)
```

```python
items["total_item_value"] = (
    items["price"]
    + items["freight_value"]
)
```

## Products

The products table contains product metadata including category, description characteristics, image count, and physical dimensions. Missing category information was handled by assigning an "Unknown" category to preserve referential integrity. Additional features such as product volume were engineered to support logistics and sales analyses.

```python
products.info()
```

```python
products.describe()
```

```python
products.head()
```

```python
products[
    products["product_category_name"].isna()
].head()
```

```python
products["product_category_name"] = (
    products["product_category_name"]
    .fillna("unknown")
)
```

```python
products["product_name_lenght"] = (
    products["product_name_lenght"]
    .fillna(0)
    .astype(int)
)
```

```python
products = products.rename(columns={
    "product_name_lenght": "product_name_length",
    "product_description_lenght": "product_description_length"
})
```

```python
payments["installment_group"] = pd.cut(
    payments["payment_installments"],
    bins=[0,1,3,6,12,24],
    labels=[
        "Single",
        "2-3",
        "4-6",
        "7-12",
        "13+"
    ]
)
```
